```python
from serial import Serial
```

```python
def get_serial_reading():
    try:
        x = Serial("COM5",9600)
        x.readline()
        y = x.readline()
        z = y.decode("utf-8")
        ans = ''
        for i in z:
            if not(i == '\r' or i== '\n'):
                ans += i
        ans = ans.split(',')
        ecg, a_x,a_y,a_z = ans
        return [ecg, a_x,a_y,a_z]
    except:
        return []
```
